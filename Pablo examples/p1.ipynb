Plotting Networks


```python
import cartopy.crs as ccrs
loading = (n.lines_t.p0.abs().mean().sort_index()/ (n.lines.s_nom_opt*n.lines.s_max_pu).sort_index()).fillna(0.)
```

```python
n.plot()
```

#PlateCarre, Mercator, Orthographic

```python
fig,ax = plt.subplots(
    figsize=(10,10),
    subplot_kw={"projection": ccrs.Mercator()}
)
n.plot(ax=ax,
       bus_colors='gray',
       branch_components=["Line"],
        # line_widths=n.lines.s_nom_opt, the results for this are 0, then this line is worthless
        line_colors=loading,
       line_cmap=plt.cm.viridis,
       color_geomap=True,
       bus_sizes=0
       )
ax.axis('off');
print(n.lines.s_nom_opt.unique())
```
